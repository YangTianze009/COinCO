```python
import pandas as pd
from pathlib import Path
```

```python
def analyze_csv(filepath):
    """分析单个CSV文件"""
    print(f"\n{'='*60}")
    print(f"📄 文件: {filepath}")
    print('='*60)
    
    df = pd.read_csv(filepath)
    
    print(f"\n📊 基本信息:")
    print(f"   行数: {len(df)}")
    print(f"   列数: {len(df.columns)}")
    
    print(f"\n📋 列名: {list(df.columns)}")
    
    print(f"\n🔍 各列详情:")
    for col in df.columns:
        print(f"\n   【{col}】")
        print(f"      dtype: {df[col].dtype}")
        print(f"      非空数: {df[col].notna().sum()} / {len(df)}")
        
        n_unique = df[col].nunique()
        print(f"      unique值数量: {n_unique}")
        
        # 如果unique值较少，显示所有；否则显示前10个样本
        if n_unique <= 15:
            print(f"      所有unique值: {df[col].unique().tolist()}")
        else:
            print(f"      前10个样本: {df[col].head(10).tolist()}")
            
        # 对于数值列，显示统计信息
        if df[col].dtype in ['int64', 'float64']:
            print(f"      min: {df[col].min()}, max: {df[col].max()}, mean: {df[col].mean():.2f}")
    
    return df
```

```python
# 分析两个CSV文件
csv_files = ['common_images.csv', 'uncommon_images.csv']

dfs = {}
for f in csv_files:
    if Path(f).exists():
        dfs[f] = analyze_csv(f)
    else:
        print(f"⚠️ 文件不存在: {f}")

# 如果两个文件都存在，比较它们
if len(dfs) == 2:
    print(f"\n{'='*60}")
    print("📊 两个文件对比")
    print('='*60)
    for name, df in dfs.items():
        print(f"   {name}: {len(df)} 行")
```
